# Tumbler Collection Data Analysis

This project explores a dataset of a personal tumbler collection.  
The analysis examines brand distribution, bottle size patterns, and limited-edition items using Python and data visualization.

Tools used:
- Python
- Pandas
- Seaborn
- Matplotlib

## Dataset Overview
The dataset contains information about each tumbler including brand, model, color, bottle size, and whether it is a limited edition item.

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace(" ", "_")
df.columns
```

```python
df["brand"].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x="brand")
plt.xticks(rotation=45)
plt.title("Distribution of Tumblers by Brand")
plt.show()
```

```python
sns.histplot(df["size_oz"], bins=10)
plt.title("Distribution of Bottle Sizes")
plt.xlabel("Size (oz)")
plt.show()
```

```python
df["limited_edition"].value_counts()
```

```python
sns.countplot(data=df, x="limited_edition")
plt.title("Limited Edition vs Standard Tumblers")
plt.show()
```

```python
sns.boxplot(data=df, x="brand", y="size_oz")
plt.xticks(rotation=45)
plt.title("Bottle Size by Brand")
plt.show()
```

## Key Findings

- The most common tumbler brand in the collection is Stanley (11 tumblers).
- The most frequent bottle size is 40 oz.
- Limited edition tumblers make up 66.7% of the collection.
- Owala and Stanley tend to have larger bottle sizes on average compared to other brands.

## Color Distribution Analysis

This section explores the most common tumbler colors in the collection.

```python
df["color"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14,6))

color_palette = {
    "Black": "black",
    "Green": "green",
    "Cherry Blossom": "#FFB7C5",
    "ELF Red": "red",
    "Sweet Hearts": "#FF6F91",
    "Pink Parade": "#FF69B4",
    "Barbie Icon": "#E0218A",
    "Drip Drop": "#4FC3F7",
    "Popsicle Pink": "#FF8DAA",
    "Blue": "#4A90E2",
    "Black Leopard": "#C0A06D",
    "Autumn Checkers": "brown",
    "Black Cherry": "#5A0F1C",
    "Rose Gold": "#B76E79",
    "Read My Lips": "#C71585",
    "Power Pink": "#FF7DB1",
    "Peony": "#F85D96",
    "Blush": "#E38C7F",
    "Dried Pine": "#4B5320",
    "Leopard": "#F85D96",
    "Blair": "#F5F5DC"
}

order = df["color"].value_counts().index
palette = [color_palette.get(c, "gray") for c in order]

sns.countplot(data=df, x="color", order=order, palette=palette)

plt.xticks(rotation=75, ha="right")
plt.title("Distribution of Tumbler Colors")
plt.tight_layout()

plt.show()
```

### Color Insights

Some colors appear more frequently than others in the collection, suggesting a preference for certain color styles. This visualization highlights which colors dominate the collection.
